# PARTE 2. SQL para Exploración de Datos (Data Discovery)

## - Se importan las librerías

```python
import pandas as pd
from sqlalchemy import create_engine
```

## - Se carga el dataset

```python
df = pd.read_csv(
    "../data/ventas_techventas_limpio.csv",
    encoding="latin-1"
)

df.head(60)
```

## - Se crea la base de datos SQLite

```python
engine = create_engine("sqlite:///../output/techventas.db")
```

## - Se Exporta el DataFrame a SQLite

```python
df.to_sql(
    "ventas",
    con=engine,
    if_exists="replace",
    index=False
)
```

## - Se Verifica que se creó correctamente

```python
pd.read_sql(
    "SELECT * FROM ventas LIMIT 5;",
    engine
)
```

## - Se crea la tabla vendedores para el JOIN

```python
vendedores = pd.DataFrame({
    "vendedor": [
        "Ana Lopez",
        "Carlos Ruiz",
        "Luis Mora",
        "Maria Torres"
    ],
    "zona": [
        "Norte",
        "Sur",
        "Norte",
        "Centro"
    ],
    "meta_mensual": [
        8000,
        7500,
        8000,
        7000
    ]
})
```

- se guarda la tabla

```python
vendedores.to_sql(
    "vendedores",
    con=engine,
    if_exists="replace",
    index=False
)
```

- Se verifica la creación de la tabla 


```python

pd.read_sql(
    "SELECT * FROM vendedores;",
    engine
)
```

## Consulta 1. Productos diferentes vendidos

```python
consulta1 = """
SELECT DISTINCT producto
FROM ventas;
"""

pd.read_sql(consulta1, engine)
```

####  Interpretación: Esta consulta permite identificar todos los productos diferentes que se encuentran registrados en el conjunto de datos.

## Consulta 2. Ventas realizadas entre enero y marzo

```python
consulta2 = """
SELECT *
FROM ventas
WHERE fecha BETWEEN '2024-01-01' AND '2024-03-31';
"""

pd.read_sql(consulta2, engine)
```

#### Interpretación: Se muestran únicamente las ventas realizadas entre el 1 de enero y el 31 de marzo de 2024.

## Consulta 3. Total de ventas por región

```python
consulta3 = """
SELECT
    region,
    SUM(cantidad * precio_unitario) AS total_vendido
FROM ventas
GROUP BY region
HAVING total_vendido > 5000;
"""

pd.read_sql(consulta3, engine)
```

#### Interpretación: Se calcula el total vendido por cada región y solo se muestran aquellas cuyo valor supera los 5000.

## Consulta 4. Estadísticas por categoría

```python
consulta4 = """
SELECT
    categoria,
    COUNT(*) AS numero_ventas,
    SUM(cantidad) AS total_unidades,
    AVG(precio_unitario) AS precio_promedio
FROM ventas
GROUP BY categoria;
"""

pd.read_sql(consulta4, engine)
```

#### Interpretación: Se obtiene un resumen por categoría con el número de ventas, la cantidad total de unidades vendidas y el precio promedio.

## Consulta 5. Información de vendedores

```python
consulta5 = """
SELECT
    v.order_id,
    v.vendedor,
    ven.zona,
    ven.meta_mensual
FROM ventas v
INNER JOIN vendedores ven
ON v.vendedor = ven.vendedor;
"""

pd.read_sql(consulta5, engine)
```

#### Interpretación: El JOIN permite relacionar la información de las ventas con los datos de cada vendedor, como su zona y la meta mensual asignada.

## Consulta 6. Productos con precio superior al promedio

```python
consulta6 = """
SELECT
    producto,
    precio_unitario
FROM ventas
WHERE precio_unitario >
(
    SELECT AVG(precio_unitario)
    FROM ventas
);
"""

pd.read_sql(consulta6, engine)
```

#### Interpretación: Se identifican los productos cuyo precio unitario es superior al promedio del catálogo.

## Conclusión

Con SQL fue posible explorar la información del conjunto de datos mediante consultas que permitieron filtrar registros, agrupar información, calcular estadísticas, relacionar tablas mediante JOIN y utilizar subconsultas. Estas herramientas facilitan el análisis de los datos y apoyan la toma de decisiones.
