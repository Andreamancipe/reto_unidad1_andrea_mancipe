# 1.3. Estadísticas con NumPy

```python
media = np.mean(precios)
mediana = np.median(precios)
desviacion = np.std(precios)
percentiles = np.percentile(precios, [25, 75])

print("Media:", media)
print("Mediana:", mediana)
print("Desviación estándar:", desviacion)
print("Percentiles:", percentiles)
```

## Interpretación de los resultados

- La media del precio unitario es de **378.25**, lo que representa el precio promedio de los productos vendidos.
